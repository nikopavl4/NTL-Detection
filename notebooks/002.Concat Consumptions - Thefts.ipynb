# Remove zero CSS_MS_HS_USE on SUCCESSOR change

```python
def mask_zeroed_sucessor(df):
    df = df.copy()
    mask = df['CSS_MS_HS_USE'] == 0
    successor_change_mask = df['SUCCESSOR'] != df['SUCCESSOR'].shift()
    rows_to_drop = mask & successor_change_mask
    df = df[~rows_to_drop]
    return df
```
